```python
from transformers import pipeline
import random
import pandas as pd

unmasker = pipeline('fill-mask', model='bert-base-cased')
```

```python
input_text = "Programatically create web application & sub site"

def transform(input_text,nb=5):
  for i in range(nb):
    orig_text_list = input_text.split()
    len_input = len(orig_text_list)
    if len_input-1>1:
      rand_idx=random.randint(1,len_input-1)
    else:
      rand_idx=0
    orig_word=orig_text_list[rand_idx]
    new_text_list = orig_text_list.copy()
    new_text_list[rand_idx] = '[MASK]'
    new_mask_sent = ' '.join(new_text_list)

    augmented_text_list = unmasker(new_mask_sent)
    """for res in augmented_text_list:
      if res['token_str'] != orig_word:
        augmented_text = res['sequence']
        break"""
    augmented_text=  augmented_text_list[0]['sequence']
    input_text=augmented_text
  return input_text
print(transform(input_text,5))
```

```python
data=[]
nb=5
i=0
with open('/kaggle/input/stackoverflow/stackoverflow_true_text.txt') as f:
    line=f.readline()
    while line:
        classe=line.split("\t")[0]
        text=line.split("\t")[1]
        print(text)
        print(i)
        i+=1
        text1=transform(text,nb)
        text2=transform(text,nb)
        dict={'Classe':classe,
          'text':text,
          'text1':text1,
          'text2':text2}
        data.append(dict)
        line = f.readline()
        if i%1000==0:
            df = pd.DataFrame(data,columns=['Classe','text','text1','text2'])
            df.to_csv('out'+str(i)+'.csv',index=False)
        

df = pd.DataFrame(data,columns=['Classe','text','text1','text2'])
df.to_csv('augmented.csv',index=False)
```
